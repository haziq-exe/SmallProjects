## Finding Euclidean distance

```python
from math import sqrt
import warnings
warnings.filterwarnings("ignore")
```

```python
def euclidean_distance(row1, row2):
 distance = 0.0
 for i in range(len(row1)):
     distance += (row1[i] - row2[i])**2
 finalDistance = sqrt(distance)
 return finalDistance   
```

```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
dfcopy = df.copy()

usefulColumns = ['show_id', 'type', 'title', 'listed_in']


Maindf = dfcopy[usefulColumns]

Maindf.tail()
```

```python
categories = df['listed_in'].str.get_dummies(sep = ', ')

Maindf = pd.concat([Maindf, categories], axis = 1)

Maindf.tail()
```

```python
dfcopy2 = Maindf.copy()
droppedColumns = ['show_id', 'type', 'title', 'listed_in']
vectorDF = dfcopy2.drop(droppedColumns, axis = 1)

vectorDF.head()

for columns in vectorDF:
    print(columns)
```

```python
def nearestNeighbours(row1, kValue):

    distances = list()
    closestN = list()
    
    for rows in range(len(vectorDF)):
        
        targetRow = vectorDF.iloc[rows]
        
        dist = euclidean_distance(row1, targetRow)
        distances.append((dist, rows))

    distances.sort(key=lambda tup: tup[0])

    for i in range(1, kValue + 1):
        closestN.append(distances[i])

    return closestN
    
```

```python
row = 0
title = dfcopy.loc[row, "title"]

while title != "Kota Factory":
    row = row + 1
    title = dfcopy.loc[row, "title"]


nearest = nearestNeighbours(vectorDF.iloc[row], 20)

similar = [t[1] for t in nearest]

for c in range(len(similar)):
    print(dfcopy.loc[similar[c], "title"])
    
```
